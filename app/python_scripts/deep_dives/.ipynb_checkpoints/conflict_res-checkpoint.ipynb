```python
# this is for the data from the three runs where we try 
# each of the conflict resolutions

# imports and reading csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data1 = pd.read_csv(r"C:\Users\allym\Documents\MQP\all_data\data_1.csv")
data3 = pd.read_csv(r"C:\Users\allym\Documents\MQP\all_data\data_3.csv")
data4 = pd.read_csv(r"C:\Users\allym\Documents\MQP\all_data\data_4.csv")
data12 = pd.read_csv(r"C:\Users\allym\Documents\MQP\all_data\data_12.csv")
```

```python
# general initializations
x=['move, exit, others', 'move, exit and reset, others', 'move, exit', 'none']
```

```python
# y1 initializations
y1=[]

avg_1_1=sum(data1['total_exit_time'])/len(data1['total_exit_time'])
y1.append(avg_1_1)

avg_1_3=sum(data3['total_exit_time'])/len(data3['total_exit_time'])
y1.append(avg_1_3)

avg_1_4=sum(data4['total_exit_time'])/len(data4['total_exit_time'])
y1.append(avg_1_4)

avg_1_12=sum(data12['total_exit_time'])/len(data12['total_exit_time'])
y1.append(avg_1_12)

# print('y1', y1)
```

```python
# y2 initializations
y2=[]

avg_2_1=sum(data1['avg_exit_time'])/len(data1['avg_exit_time'])
y2.append(avg_2_1)

avg_2_3=sum(data3['avg_exit_time'])/len(data3['avg_exit_time'])
y2.append(avg_2_3)

avg_2_4=sum(data4['avg_exit_time'])/len(data4['avg_exit_time'])
y2.append(avg_2_4)

avg_2_12=sum(data12['avg_exit_time'])/len(data12['avg_exit_time'])
y2.append(avg_2_12)

# print('y2', y2)
```

```python
# y3 initializations
y3=[]

avg_3_1=sum(data1['avg_collisions_total'])/len(data1['avg_collisions_total'])
y3.append(avg_3_1)

avg_3_3=sum(data3['avg_collisions_total'])/len(data3['avg_collisions_total'])
y3.append(avg_3_3)

avg_3_4=sum(data4['avg_collisions_total'])/len(data4['avg_collisions_total'])
y3.append(avg_3_4)

avg_3_12=sum(data12['avg_collisions_total'])/len(data12['avg_collisions_total'])
y3.append(avg_3_12)

# print('y3', y3)
```

```python
# y4 initializations
y4=[]

avg_4_1=sum(data1['total_avg_occ_all_time'])/len(data1['total_avg_occ_all_time'])
y4.append(avg_4_1)

avg_4_3=sum(data3['total_avg_occ_all_time'])/len(data3['total_avg_occ_all_time'])
y4.append(avg_4_3)

avg_4_4=sum(data4['total_avg_occ_all_time'])/len(data4['total_avg_occ_all_time'])
y4.append(avg_4_4)

avg_4_12=sum(data12['total_avg_occ_all_time'])/len(data12['total_avg_occ_all_time'])
y4.append(avg_4_12)

# print('y4', y4)
```

```python
# y5 initializations
y5=[]

avg_5_1=sum(data1['evaluation_metric'])/len(data1['evaluation_metric'])
y5.append(avg_5_1)

avg_5_3=sum(data3['evaluation_metric'])/len(data3['evaluation_metric'])
y5.append(avg_5_3)

avg_5_4=sum(data4['evaluation_metric'])/len(data4['evaluation_metric'])
y5.append(avg_5_4)

avg_5_12=sum(data12['evaluation_metric'])/len(data12['evaluation_metric'])
y5.append(avg_5_12)

# print('y5', y5)
```

```python
# building the bar graphs

# plot 1
plt.bar(x,y1)
plt.title('Average total exit time based on conflict strategy')
plt.xlabel('Conflict resolution strategy')
plt.ylabel('Average total exit time')
plt.show()

# plot 2
plt.bar(x,y2)
plt.title('Mean of average exit time based on conflict strategy')
plt.xlabel('Conflict resolution strategy')
plt.ylabel('Mean of average exit time')
plt.show()

# plot 3
plt.bar(x,y3)
plt.title('Mean of average collisions based on conflict strategy')
plt.xlabel('Conflict resolution strategy')
plt.ylabel('Mean of average collisions')
plt.show()

# plot 4
plt.bar(x,y4)
plt.title('Mean of average area occupancy based on conflict strategy')
plt.xlabel('Conflict resolution strategy')
plt.ylabel('Mean of average area occupancy')
plt.show()

# plot 5
plt.bar(x,y5)
plt.title('Average evaluation metric based on conflict strategy')
plt.xlabel('Conflict resolution strategy')
plt.ylabel('Average evaluation metric')
plt.show()

print('y1',y1)
print('y2',y2)
print('y3',y3)
print('y4',y4)
print('y5',y5)
```
