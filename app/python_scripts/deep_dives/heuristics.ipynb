```python
# this is for the data from the three runs where we try 
# each of the heuristic functions

# imports and reading csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec


data9 = pd.read_csv(r"C:\Users\allym\Documents\MQP\all_data\data_9.csv")
data10 = pd.read_csv(r"C:\Users\allym\Documents\MQP\all_data\data_10.csv")
data11 = pd.read_csv(r"C:\Users\allym\Documents\MQP\all_data\data_11.csv")
```

```python
# general initializations
x=['Euclidean', 'Diagonal', 'Manhattan']
```

```python
# y1 initializations
y1=[]
y1e=[]

avg_1_9=np.mean(data9['total_exit_time'])
avg_1_10=np.mean(data10['total_exit_time'])
avg_1_11=np.mean(data11['total_exit_time'])

y1.append(avg_1_9)
y1.append(avg_1_10)
y1.append(avg_1_11)

y1e.append(np.std(data9['total_exit_time']))
y1e.append(np.std(data10['total_exit_time']))
y1e.append(np.std(data11['total_exit_time']))

# print('y1', y1)
```

```python
# y2 initializations
y2=[]
y2e=[]

avg_2_9=np.mean(data9['avg_exit_time'])
avg_2_10=np.mean(data10['avg_exit_time'])
avg_2_11=np.mean(data11['avg_exit_time'])

y2.append(avg_2_9)
y2.append(avg_2_10)
y2.append(avg_2_11)

y2e.append(np.std(data9['avg_exit_time']))
y2e.append(np.std(data10['avg_exit_time']))
y2e.append(np.std(data11['avg_exit_time']))

# print('y2', y2)
```

```python
# y3 initializations
y3=[]
y3e=[]

avg_3_9=np.mean(data9['avg_collisions_total'])
avg_3_10=np.mean(data10['avg_collisions_total'])
avg_3_11=np.mean(data11['avg_collisions_total'])

y3.append(avg_3_9)
y3.append(avg_3_10)
y3.append(avg_3_11)

y3e.append(np.std(data9['avg_collisions_total']))
y3e.append(np.std(data10['avg_collisions_total']))
y3e.append(np.std(data11['avg_collisions_total']))

# print('y3', y3)
```

```python
# y4 initializations
y4=[]
y4e=[]

avg_4_9=np.mean(data9['total_avg_occ_all_time'])
avg_4_10=np.mean(data10['total_avg_occ_all_time'])
avg_4_11=np.mean(data11['total_avg_occ_all_time'])

y4.append(avg_4_9)
y4.append(avg_4_10)
y4.append(avg_4_11)

y4e.append(np.std(data9['total_avg_occ_all_time']))
y4e.append(np.std(data10['total_avg_occ_all_time']))
y4e.append(np.std(data11['total_avg_occ_all_time']))

# print('y4', y4)
```

```python
# y5 initializations
y5=[]
y5e=[]

avg_5_9=np.mean(data9['evaluation_metric'])
avg_5_10=np.mean(data10['evaluation_metric'])
avg_5_11=np.mean(data11['evaluation_metric'])

y5.append(avg_5_9)
y5.append(avg_5_10)
y5.append(avg_5_11)

y5e.append(np.std(data9['evaluation_metric']))
y5e.append(np.std(data10['evaluation_metric']))
y5e.append(np.std(data11['evaluation_metric']))

# print('y4', y4)
```

```python
# building the bar graphs

fig=plt.figure(figsize=(30, 12))
gs = GridSpec(nrows=2, ncols=3)
ax0 = fig.add_subplot(gs[0, 0])
plt.bar(x,y1,yerr=y1e, align='center', alpha=0.5, ecolor='black', capsize=10)
plt.title('Average Total Exit Time Based on Heuristic')
plt.xlabel('Heuristic')
plt.ylabel('Average Total Exit Time')
ax1 = fig.add_subplot(gs[1, 0])
plt.bar(x,y2,yerr=y2e, align='center', alpha=0.5, ecolor='black', capsize=10)
plt.title('Mean of Average Exit Time Based on Heuristic')
plt.xlabel('Heuristic')
plt.ylabel('Mean of Average Exit Time')
ax2 = fig.add_subplot(gs[0, 1])
plt.bar(x,y3,yerr=y3e, align='center', alpha=0.5, ecolor='black', capsize=10)
plt.title('Mean of Average Collisions Based on Heuristic')
plt.xlabel('Heuristic')
plt.ylabel('Mean of Average Collisions')
ax3 = fig.add_subplot(gs[1, 1])
plt.bar(x,y4,yerr=y4e, align='center', alpha=0.5, ecolor='black', capsize=10)
plt.title('Mean of Average Area Occupancy Based on Heuristic')
plt.xlabel('Heuristic')
plt.ylabel('Mean of Average Area Occupancy')
ax3 = fig.add_subplot(gs[:, 2])
plt.bar(x,y5,yerr=y5e, align='center', alpha=0.5, ecolor='black', capsize=10)
plt.title('Average Evaluation Metric Based on Heuristic')
plt.xlabel('Heuristic')
plt.ylabel('Average Evaluation Metric')
```
