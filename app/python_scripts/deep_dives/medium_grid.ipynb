```python
# this is for the data from the large grid runs ()

# imports and reading csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec

data15 = pd.read_csv(r"C:\Users\allym\Documents\MQP\all_data\data_19.csv")
data16 = pd.read_csv(r"C:\Users\allym\Documents\MQP\all_data\data_20.csv")
data17 = pd.read_csv(r"C:\Users\allym\Documents\MQP\all_data\data_21.csv")
```

```python
# general initializations
# x=['10,10,10,0', '25,25,25,15']
x=[10,25,50]
```

```python
# y1 initializations
y1=[]
y1e=[]

avg_1_15=np.mean(data15['total_exit_time'])
avg_1_16=np.mean(data16['total_exit_time'])
avg_1_17=np.mean(data17['total_exit_time'])

y1.append(avg_1_15)
y1.append(avg_1_16)
y1.append(avg_1_17)

y1e.append(np.std(data15['total_exit_time']))
y1e.append(np.std(data16['total_exit_time']))
y1e.append(np.std(data17['total_exit_time']))

# print('y1', y1)
```

```python
# y2 initializations
y2=[]
y2e=[]

avg_2_15=np.mean(data15['avg_exit_time'])
avg_2_16=np.mean(data16['avg_exit_time'])
avg_2_17=np.mean(data17['avg_exit_time'])

y2.append(avg_2_15)
y2.append(avg_2_16)
y2.append(avg_2_17)

y2e.append(np.std(data15['avg_exit_time']))
y2e.append(np.std(data16['avg_exit_time']))
y2e.append(np.std(data17['avg_exit_time']))

# print('y2', y2)
```

```python
# y3 initializations
y3=[]
y3e=[]

avg_3_15=np.mean(data15['avg_collisions_total'])
avg_3_16=np.mean(data16['avg_collisions_total'])
avg_3_17=np.mean(data17['avg_collisions_total'])

y3.append(avg_3_15)
y3.append(avg_3_16)
y3.append(avg_3_17)

y3e.append(np.std(data15['avg_collisions_total']))
y3e.append(np.std(data16['avg_collisions_total']))
y3e.append(np.std(data17['avg_collisions_total']))

# print('y3', y3)
```

```python
# y4 initializations
y4=[]
y4e=[]

avg_4_15=np.mean(data15['total_avg_occ_all_time'])
avg_4_16=np.mean(data16['total_avg_occ_all_time'])
avg_4_17=np.mean(data17['total_avg_occ_all_time'])

y4.append(avg_4_15)
y4.append(avg_4_16)
y4.append(avg_4_17)

y4e.append(np.std(data15['total_avg_occ_all_time']))
y4e.append(np.std(data16['total_avg_occ_all_time']))
y4e.append(np.std(data17['total_avg_occ_all_time']))

# print('y4', y4)
```

```python
# y5 initializations
y5=[]
y5e=[]

avg_5_15=np.mean(data15['evaluation_metric'])
avg_5_16=np.mean(data16['evaluation_metric'])
avg_5_17=np.mean(data17['evaluation_metric'])

y5.append(avg_5_15)
y5.append(avg_5_16)
y5.append(avg_5_17)

y5e.append(np.std(data15['evaluation_metric']))
y5e.append(np.std(data16['evaluation_metric']))
y5e.append(np.std(data17['evaluation_metric']))

# print('y5', y5)
```

```python
# # building the bar graphs

fig=plt.figure(figsize=(30, 12))
gs = GridSpec(nrows=2, ncols=3)

# plot 1
ax0 = fig.add_subplot(gs[0, 0])
plt.scatter(x,y1,s=75)
plt.errorbar(x,y1,yerr=y1e)
plt.title('Average Total Exit Time Based on Num Peds')
plt.xlabel('Number of Pedestrians')
plt.ylabel('Average Total Exit Time')

# plot 2
ax1 = fig.add_subplot(gs[1, 0])
plt.scatter(x,y2,s=75)
plt.errorbar(x,y2,yerr=y2e)
plt.title('Mean of Average Exit Time Based on Num Peds')
plt.xlabel('Number of Pedestrians')
plt.ylabel('Mean of Average Exit Time')

# plot 3
ax2 = fig.add_subplot(gs[0, 1])
plt.scatter(x,y3,s=75)
plt.errorbar(x,y3,yerr=y3e)
plt.title('Mean of Average Collisions Based on Num Peds')
plt.xlabel('Number of Pedestrians')
plt.ylabel('Mean of Average Collisions')

# plot 4
ax3 = fig.add_subplot(gs[1, 1])
plt.scatter(x,y4,s=75)
plt.errorbar(x,y4,yerr=y4e)
plt.title('Mean of Average Area Occupancy Based on Num Peds')
plt.xlabel('Number of Pedestrians')
plt.ylabel('Mean of Average Area Occupancy')

# plot 5
ax3 = fig.add_subplot(gs[:, 2])
plt.scatter(x,y5,s=75)
plt.errorbar(x,y5,yerr=y5e)
plt.title('Average Evaluation Metric Based on Num Peds')
plt.xlabel('Number of Pedestrians')
plt.ylabel('Average Evaluation Metric')
```
